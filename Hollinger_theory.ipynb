```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

### Adding reference of each player according to Hollinger's theory

```python
#Read our data to a Dataframe
stats_df = pd.read_csv('NBA_clean_data.csv')
hollinger_theory = pd.read_csv('hollinger_theory.csv')
```

```python
#This function will add the reference of each player according to Hollinger's theory.
def addHollingerTheoryReference(df):
    df_copy=df.copy()
    df_copy['ht']= ''
    for index,row in df_copy.iterrows():
        if(row['PER'] >= 35.0):
            df_copy.at[index,'ht'] = hollinger_theory['Reference guide'][0]
        elif ((row['PER'] >= 30.0) & (row['PER'] < 35.0) ):
            df_copy.at[index,'ht'] = hollinger_theory['Reference guide'][1]
        elif ((row['PER'] >= 27.5) & (row['PER'] < 30.0) ):
            df_copy.at[index,'ht'] = hollinger_theory['Reference guide'][2]
        elif ((row['PER'] >= 25) & (row['PER'] < 27.5) ):
            df_copy.at[index,'ht'] = hollinger_theory['Reference guide'][3]
        elif ((row['PER'] >= 22.5) & (row['PER'] < 27.5) ):
            df_copy.at[index,'ht'] = hollinger_theory['Reference guide'][4]
        elif ((row['PER'] >= 20.0) & (row['PER'] < 22.5) ):
            df_copy.at[index,'ht'] = hollinger_theory['Reference guide'][5]
        elif ((row['PER'] >= 18.0) & (row['PER'] < 20.0) ):
            df_copy.at[index,'ht'] = hollinger_theory['Reference guide'][6]
        elif ((row['PER'] >= 16.5) & (row['PER'] < 18.0) ):
            df_copy.at[index,'ht'] = hollinger_theory['Reference guide'][7]
        elif ((row['PER'] >= 15.0) & (row['PER'] < 16.5) ):
            df_copy.at[index,'ht'] = hollinger_theory['Reference guide'][8]
        elif ((row['PER'] >= 13.0) & (row['PER'] < 15.0) ):
            df_copy.at[index,'ht'] = hollinger_theory['Reference guide'][9]
        elif ((row['PER'] >= 11.0) & (row['PER'] < 13.0) ):
            df_copy.at[index,'ht'] = hollinger_theory['Reference guide'][10]
        elif ((row['PER'] >= 9.0) & (row['PER'] < 11.0) ):
            df_copy.at[index,'ht'] = hollinger_theory['Reference guide'][11]
        else:
            df_copy.at[index,'ht'] = hollinger_theory['Reference guide'][12] 
    return df_copy
```

```python
# We will add a column to our clean big data with the reference of each player.
Combined_df = addHollingerTheoryReference(stats_df)
Combined_df
```

```python
#Create CSV file of our Hollinger theory combined with the big stats dataset
Combined_df.to_csv('NBA_Hollinger_Data.csv', index=False)
```
