```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Read our data to a Dataframe
final_df = pd.read_csv('NBA_Hollinger_Data.csv')
mvp_df = pd.read_csv('NBA_MVP_clean_data.csv')
```

### Visualize Hollinger theory

```python
plt.figure(figsize=(35,20))
final_df['ht'].value_counts().plot(kind='pie',labels=None,autopct='%1.1f%%',textprops= {'fontsize':20})
plt.legend(fontsize=12,loc="upper right",labels=final_df['ht'])
```

### In this pie chart we can see the precentge of players in our dataset for each category by Hollinger's theory.

### Visualize MVP

```python
#Combine the 2 dataframes into 1 DataFrame
top_per = final_df.loc[final_df.groupby('Year')['PER'].idxmax()]
top_per = top_per.sort_values(by='Year',ascending=False)

top_per.insert(0,'MVP_This_Year',mvp_df['Player'].tolist())
top_per.reset_index()
top_per['Player'] = top_per['Player'].str.replace('*', '')
match = pd.get_dummies(top_per['Player']==top_per['MVP_This_Year'], drop_first=True)
top_per.insert(0,'We_Have_A_Match',match)

```

### We will examine the relationship between the player with the highest PER and the MVP of that season

```python
sns.scatterplot(data=top_per, y="Year", x="We_Have_A_Match")
```

### We notice that as we go up in years the connection between the PER and the MVP is stronger

### We will check what percentage of MVP players have the highest PER of their season in the last 30 years

```python
plt.figure(figsize=(10,10))
plt.pie(x=top_per['We_Have_A_Match'].value_counts(),labels=None,autopct='%1.1f%%',textprops= {'fontsize':20})
plt.legend(fontsize=14,loc="upper right",labels=['The MVP does not have the highest PER','The MVP does have the highest PER'])
```
